```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Nous raisonnons ici par heure de vol, puis, en utilisant la vitesse de croisière, la traduisons en distance parcourue.

Fonction pour définir l'amortissement par km

```python
def amortissement_par_km (prix_avion, duree_amortissement, nbr_heures_vol_annuel, vitesse_croisiere_avion):
    return prix_avion/(duree_amortissement*nbr_heures_vol_annuel*vitesse_croisiere_avion)
```

Fonction pour définir le salaire PNT par km

```python
def salaires_PNT_par_km (salaire_PNT_horaire, nbr_PNT, vitesse_croisiere_avion):
    return nbr_PNT*salaire_PNT_horaire/vitesse_croisiere_avion
```

Fonction pour définir le salaire PNC par km

```python
def salaires_PNC_par_km (salaire_PNC_horaire, nbr_PNC, vitesse_croisiere_avion):
    return nbr_PNC*salaire_PNC_horaire/vitesse_croisiere_avion
```


Fonction pour estimer l'assurance par km

```python
def assurance_par_km (prix_assurance_annuel, nbr_heures_vol_annuel, vitesse_croisiere_avion):
    return prix_assurance_annuel/(nbr_heures_vol_annuel*vitesse_croisiere_avion)
```

Fonction pour redevance de stationnement par km

```python
def redevance_de_stationnement_par_km (redevance_de_stationnement_horaire, vitesse_croisiere_avion):
    return redevance_de_stationnement_horaire/vitesse_croisiere_avion
```

Fonction pour actifs aeroports par km 

```python
def actifs_aeroports_par_km (prix_actifs_aeroports_annuel, nbr_heures_vol_annuel, vitesse_croisiere_avion):
    return prix_actifs_aeroports_annuel/(nbr_heures_vol_annuel*vitesse_croisiere_avion)
```

Fonction pour ticketing par km

```python
def ticketing_par_km (ticketing_annuel, nbr_heures_vol_annuel, vitesse_croisiere_avion):
    return ticketing_annuel/(nbr_heures_vol_annuel*vitesse_croisiere_avion)
```

Fonction pour coûts administratifs par km

```python
def couts_administratif_par_km (couts_administratif_annuel, nbr_heures_vol_annuel, vitesse_croisiere_avion):
    return couts_administratif_annuel/(nbr_heures_vol_annuel*vitesse_croisiere_avion)
```





Fonction pour coût de l'énergie thermique par km

```python
def cout_energie_thermique_par_km (consommation_energie_thermique,cout_energie_thermique) :
    return consommation_energie_thermique*cout_energie_thermique   #L/km * €/L
```

Fonction pour coût de l'énergie électrique par km

```python
def cout_energie_electrique_par_km (consommation_energie_electrique,cout_energie_electrique) :
    return consommation_energie_electrique*cout_energie_electrique  #kWh/km * €/kWh
```

Fonction pour maintenance par km

```python
def maintenance_par_km (maintenance_annuel, nbr_heures_vol_annuel, vitesse_croisiere_avion):
    return maintenance_annuel/(nbr_heures_vol_annuel*vitesse_croisiere_avion)
```

Fonction pour taxes de redevances aeroportuaires par km 

```python
def taxes_redevances_aeroportuaires_par_km (taxes_redevances_aeroportuaires_annuel, nbr_heures_vol_annuel, vitesse_croisiere_avion):
    return taxes_redevances_aeroportuaires_annuel/(nbr_heures_vol_annuel*vitesse_croisiere_avion)
```

Fonction pour cout turnaround par km

```python
def cout_turnaround_par_km (cout_turnaround_annuel, nbr_heures_vol_annuel, vitesse_croisiere_avion):
    return cout_turnaround_annuel/(nbr_heures_vol_annuel*vitesse_croisiere_avion)
```


Fonction pour estimer coûts fixes

```python
def estimer_couts_fixes_par_km(amortissement_avion_par_km, salaires_PNT_par_km, salaire_PNC_par_km, assurances_par_km, redevance_de_stationnement_par_km, actifs_aeroports_par_km, ticketing_par_km, cout_administratif_par_km):
    return amortissement_avion_par_km + salaires_PNT_par_km + salaire_PNC_par_km + assurances_par_km + redevance_de_stationnement_par_km + actifs_aeroports_par_km + ticketing_par_km + cout_administratif_par_km
```


Fonction pour estimer les coûts variables par km

```python
def estimer_couts_variables_par_km(cout_energie_par_km, maintenance_par_km, taxes_redevances_aeroportuaires_par_km, cout_turnaround_par_km):
    return cout_energie_par_km + maintenance_par_km + taxes_redevances_aeroportuaires_par_km + cout_turnaround_par_km
```

Fonction pour calculer le CASK

```python
def calculer_cask(cout_fixe_par_km, cout_variable_par_km, distance, nombre_sieges):
    cout_total_vol_par_km = cout_fixe_par_km + cout_variable_par_km
    return cout_total_vol_par_km / nombre_sieges 
```

Fonction pour calculer le RASK :

```python
def calculer_rask(prix_billet, distance, taux_remplissage):
    revenus_total = prix_billet * taux_remplissage
    return revenus_total / distance 
```


Fonction rentabilité

```python
def calculer_rentabilite(distance, amortissement_avion_par_km, salaires_PNT_par_km, salaire_PNC_par_km, assurances_par_km, redevance_de_stationnement_par_km, actifs_aeroports_par_km, ticketing_par_km, cout_administratif_par_km, cout_energie_par_km, maintenance_par_km, taxes_redevances_aeroportuaires_par_km, cout_turnaround_par_km, prix_billet, nombre_sieges, taux_remplissage):
    cout_fixe_par_km = estimer_couts_fixes_par_km(amortissement_avion_par_km, salaires_PNT_par_km, salaire_PNC_par_km, assurances_par_km, redevance_de_stationnement_par_km, actifs_aeroports_par_km, ticketing_par_km, cout_administratif_par_km)
    cout_variable_par_km = estimer_couts_variables_par_km(cout_energie_par_km, maintenance_par_km, taxes_redevances_aeroportuaires_par_km, cout_turnaround_par_km)
    cask = calculer_cask(cout_fixe_par_km, cout_variable_par_km, distance, nombre_sieges)
    rask = calculer_rask(prix_billet, distance,taux_remplissage)
    rentabilite = rask - cask
    return cask, rask, rentabilite
```

Tracé

```python
# Liste pour stocker les résultats
prix_billets = list(range(5, 200))  # Variation du prix du billet de 10€ à 200€
rentabilites = []

# Calcul de la rentabilité pour chaque prix de billet
for prix_billet in prix_billets:
    _, _, rentabilite = calculer_rentabilite(distance, amortissement_avion_par_km, salaires_PNT_par_km, salaire_PNC_par_km, assurances_par_km, redevance_de_stationnement_par_km, actifs_aeroports_par_km, ticketing_par_km, cout_administratif_par_km, cout_energie_par_km, maintenance_par_km, taxes_redevances_aeroportuaires_par_km, cout_turnaround_par_km, prix_billet, nombre_sieges, taux_remplissage)
    rentabilites.append(rentabilite)


# Trouver le prix du billet pour lequel la rentabilité est nulle
prix_rentable = None
for i in range(len(rentabilites) - 1):
    if rentabilites[i] <= 0 <= rentabilites[i + 1]:
        prix_rentable = prix_billets[i]
        break

# Tracé de la courbe améliorée
plt.figure(figsize=(12, 8))
plt.plot(prix_billets, rentabilites, marker='o', linestyle='-', color='b', label='Rentabilité')
plt.title('Rentabilité en fonction du prix du billet', fontsize=16)
plt.xlabel('Prix du billet (€)', fontsize=14)
plt.ylabel('Rentabilité (€ par siège-kilomètre)', fontsize=14)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.axhline(0, color='black', linewidth=1)
if prix_rentable is not None:
    plt.axvline(prix_rentable, color='red', linestyle='--', linewidth=1, label=f'Prix de rentabilité: {prix_rentable}€')
    plt.annotate(f'Prix de rentabilité\n({prix_rentable}€)', 
                 xy=(prix_rentable, 0), 
                 xytext=(prix_rentable + 5, 0.05),
                 arrowprops=dict(facecolor='black', arrowstyle='->'),
                 fontsize=12)

plt.legend()
plt.show()
```
